```python
speeddating = speeddating.drop(columns=['quantos_matches_acho_terei', 'quantos_acho_vao_gostar_demim'])
```

```python
speeddating.isnull().sum()
```

```python
speeddating['avaliacao_interessescomuns'] = speeddating['avaliacao_interessescomuns'].replace(["F"], '0')
speeddating['avaliacao_interessescomuns'] = speeddating['avaliacao_interessescomuns'].astype(float)
speeddating['avaliacao_interessescomuns'].fillna((speeddating['avaliacao_interessescomuns'].mean()), inplace=True)
```

```python
speeddating['correlacao_interesse'] = speeddating['correlacao_interesse'].str.slice(0)
```

```python
speeddating['idade'].fillna((speeddating['idade'].mean()), inplace=True)
speeddating['idade_parceiro'].fillna((speeddating['idade_parceiro'].mean()), inplace=True)
speeddating['parceiro_gostou_demim_acho'].fillna((speeddating['parceiro_gostou_demim_acho'].mean()), inplace=True)
speeddating['conhecia_parceiro'].fillna((speeddating['conhecia_parceiro'].mean()), inplace=True)
speeddating['parceiro_sincero'].fillna((speeddating['parceiro_sincero'].mean()), inplace=True)
speeddating['parceiro_ambicioso'].fillna((speeddating['parceiro_ambicioso'].mean()), inplace=True)
speeddating['interessescomunsparceiro'].fillna((speeddating['interessescomunsparceiro'].mean()), inplace=True)
```

```python
speeddating[['pref_parceiro_engracado']].boxplot()

plt.figure(figsize=(8,10))
g1 = sns.pointplot(x='idade', y='pref_parceiro_atratividade', 
                   data=speeddating)
g1.set_xticklabels(g1.get_xticklabels(),rotation=90)
g1.set_title("Idade Pessoa Entrevistada e Importancia de Atratividade", fontsize=15)
g1.set_xlabel("Idade Pessoa Entrevistada")
g1.set_ylabel("Atratividade", fontsize=12)
plt.show()
```

```python
speeddating['genero'] = speeddating['genero'].str.replace("b", "")
speeddating['genero'] = speeddating['genero'].str.replace("'", "")

speeddating['raca'] = speeddating['raca'].str.replace("b", "")
speeddating['raca'] = speeddating['raca'].str.replace("'", "")

speeddating['raca_parceiro'] = speeddating['raca_parceiro'].str.replace("b", "")
speeddating['raca_parceiro'] = speeddating['raca_parceiro'].str.replace("'", "")

speeddating['mesma_raca'] = speeddating['mesma_raca'].str.replace("b", "")
speeddating['mesma_raca'] = speeddating['mesma_raca'].str.replace("'", "")

speeddating['formacao'] = speeddating['formacao'].str.replace("b", "")
speeddating['formacao'] = speeddating['formacao'].str.replace("'", "")

speeddating['decisao'] = speeddating['decisao'].replace(["b'1'"], 'sim') 
speeddating['decisao'] = speeddating['decisao'].replace(["b'0'"], 'nao')

speeddating['decisao_parceiro'] = speeddating['decisao_parceiro'].replace(["b'1'"], 'sim') 
speeddating['decisao_parceiro'] = speeddating['decisao_parceiro'].replace(["b'0'"], 'nao')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

speeddating[['participantes', 'idade']] = scaler.fit_transform(speeddating[['participantes', 'idade']])

speeddating[['idade_parceiro']] = scaler.fit_transform(speeddating[['idade_parceiro']])

speeddating[['importancia_mesma_raca', 'importancia_mesma_religiao']] = scaler.fit_transform(speeddating[['importancia_mesma_raca', 'importancia_mesma_religiao']])

speeddating[['pref_parceiro_atratividade', 'pref_parceiro_sinceridade']] = scaler.fit_transform(speeddating[['pref_parceiro_atratividade', 'pref_parceiro_sinceridade']])

speeddating[['pref_parceiro_inteligencia', 'pref_parceiro_engracado']] = scaler.fit_transform(speeddating[['pref_parceiro_inteligencia', 'pref_parceiro_engracado']])

speeddating[['pref_parceiro_ambicao', 'pref_parceiro_interessescomuns']] = scaler.fit_transform(speeddating[['pref_parceiro_ambicao', 'pref_parceiro_interessescomuns']])

speeddating[['avaliacao_atratividade', 'avaliacao_sinceridade']] = scaler.fit_transform(speeddating[['avaliacao_atratividade', 'avaliacao_sinceridade']])
```

```python
speeddating[['busco_inteligencia', 'busco_engracado']] = scaler.fit_transform(speeddating[['busco_inteligencia', 'busco_engracado']])

speeddating[['busco_ambicao', 'busco_interessescomuns']] = scaler.fit_transform(speeddating[['busco_ambicao', 'busco_interessescomuns']])

speeddating[['sou_atrativo', 'sou_sincero']] = scaler.fit_transform(speeddating[['sou_atrativo', 'sou_sincero']])

speeddating[['sou_ambicioso', 'parceiro_atrativo']] = scaler.fit_transform(speeddating[['sou_ambicioso', 'parceiro_atrativo']])

speeddating[['parceiro_sincero', 'parceiro_inteligente']] = scaler.fit_transform(speeddating[['parceiro_sincero', 'parceiro_inteligente']])

speeddating[['parceiro_engracado', 'parceiro_ambicioso']] = scaler.fit_transform(speeddating[['parceiro_engracado', 'parceiro_ambicioso']])

speeddating[['interessescomunsparceiro', 'esportes']] = scaler.fit_transform(speeddating[['interessescomunsparceiro', 'esportes']])

speeddating[['esportes_tv', 'exercicio']] = scaler.fit_transform(speeddating[['esportes_tv', 'exercicio']])

speeddating[['jantares', 'museus']] = scaler.fit_transform(speeddating[['jantares', 'museus']])

speeddating[['arte', 'trilhas']] = scaler.fit_transform(speeddating[['arte', 'trilhas']])

speeddating[['videogames', 'leitura']] = scaler.fit_transform(speeddating[['videogames', 'leitura']])

speeddating[['tv', 'teatro']] = scaler.fit_transform(speeddating[['tv', 'teatro']])

speeddating[['filmes', 'shows']] = scaler.fit_transform(speeddating[['filmes', 'shows']])

speeddating[['musica', 'shopping']] = scaler.fit_transform(speeddating[['musica', 'shopping']])

speeddating[['yoga', 'correlacao_interesses']] = scaler.fit_transform(speeddating[['yoga', 'correlacao_interesses']])

speeddating[['expectativas', 'gostou_parceiro']] = scaler.fit_transform(speeddating[['expectativas', 'gostou_parceiro']])

speeddating[['parceiro_gostou_demim_acho', 'conhecia_parceiro']] = scaler.fit_transform(speeddating[['parceiro_gostou_demim_acho', 'conhecia_parceiro']])
```

```python
speeddating[['avaliacao_inteligencia']] = scaler.fit_transform(speeddating[['avaliacao_inteligencia']])

speeddating[['avaliacao_ambicao', 'avaliacao_interessescomuns']] = scaler.fit_transform(speeddating[['avaliacao_ambicao', 'avaliacao_interessescomuns']])

speeddating[['busco_atratividade', 'busco_sinceridade']] = scaler.fit_transform(speeddating[['busco_atratividade', 'busco_sinceridade']])
```
